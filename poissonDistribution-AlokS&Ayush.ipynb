# FAQs

**Q1  What is a Poisson distribution?**

A Poisson distribution is defined as a **discrete frequency distribution** that gives the **probability of the number of independent events** that **occur in the fixed time.**
OR
A discrete Random variable x which has the following Probability mass function: P(x) = [(Lambda)^x * (e)^(-Lambda)] / x!, x=0,1,2,3...infinite 
Or
It is a probability distribution that measures *how many times* an *event is likely to occur* within a *specified period of time*.
**E.g:**
- No. of accidents occurring in a city from 6 pm to 10 pm
- No. of patients arriving in the Emergency Room between 10 pm to 12 pm

**Q2  When do we use Poisson distribution?**

Poisson distribution is used when the **independent events** occurring at a **constant rate** within the given interval of time are provided.

**Q3  What is the difference between the Poisson distribution and normal distribution?**

The major difference between the Poisson distribution and the normal distribution is that the *Poisson distribution is discrete* whereas the *normal distribution is continuous.* If the **mean of the Poisson distribution becomes larger**, then the **Poisson distribution is similar to the normal distribution.**

**Q4  Are the mean and variance of the Poisson distribution the same?**

The mean and the variance of the Poisson distribution are the same, which is equal to the average number of successes that occur in the given interval of time.

**Q5  Mention the three important constraints in Poisson distribution.**

The three important constraints used in Poisson distribution are:
- The number of trials (n) tends to infinity
- The probability of success (p) tends to zero 
- np=1, which is finite.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Accidents Per Hour---x
accidents_per_hour = np.array([10, 8, 12, 15, 9, 11, 14, 13, 7, 10, 16, 11, 10, 9, 12, 8, 15, 11, 10, 13])

# Mean----Lambda
mean_accidents_per_hour = np.mean(accidents_per_hour)

# Poisson distribution 
poisson_dist = poisson(mean_accidents_per_hour)


# Random Samples from distribution
simulated_data = poisson_dist.rvs(size=1000)

# Visualize the Poisson distribution
plt.figure(figsize=(10, 6))
sns.distplot(simulated_data, kde=True, hist=False, color='skyblue')
plt.xlabel('Number of Accidents')
plt.ylabel('Probability')
plt.xticks(np.arange(0, 25))
plt.grid(True)
plt.show()


```

```python
from scipy.stats import norm
std_accidents_per_hour = np.std(accidents_per_hour)
# Generate Normal distribution
normal_dist = norm(mean_accidents_per_hour, std_accidents_per_hour)

# Generate Standard Normal distribution
standard_normal_dist = norm(0, 1)

x_values = np.arange(0, 25)

# Probability
probabilities_normal = normal_dist.pdf(x_values)
probabilities_standard_normal = standard_normal_dist.pdf(x_values)

# Plots
plt.figure(figsize=(12, 8))
# Normal Distribution
plt.subplot(1, 2, 1)
plt.plot(x_values, probabilities_normal, color='green')
plt.title('Normal Distribution')
plt.xlabel('Number of Accidents')
plt.ylabel('Probability')
plt.grid(True)
# Standard Normal Distribution
plt.subplot(1, 2, 2)
plt.plot(x_values, probabilities_standard_normal, color='orange')
plt.title('Standard Normal Distribution')
plt.xlabel('Number of Accidents')
plt.ylabel('Probability')

plt.grid(True)
plt.show()
```

# Call Example

```python
from scipy.stats import poisson
import numpy as np
# Example: Customer Arrivals at a Call Center
lambda_parameter = 10  # Average number of calls per hour
time_period = 1       # Time period in hours
 
# Generate x values (number of calls) from 0 to 30
x_values = np.arange(0, 31)
 
# Calculate Poisson PMF for each x value
poisson_pmf = poisson.pmf(x_values, lambda_parameter * time_period)
 
# Visualize the Poisson distribution
plt.bar(x_values, poisson_pmf, label='Poisson (=10)')
plt.title('Customer Arrivals at a Call Center (Poisson Distribution)')
plt.xlabel('Number of Calls in 1 Hour')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
