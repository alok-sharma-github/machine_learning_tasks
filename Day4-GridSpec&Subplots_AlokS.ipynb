# GridSpec and Subplots

```python
# Subplots

# Data
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (subplt1, subplt2) = plt.subplots(1, 2)
# Plot on the first subplot
subplt1.plot(x, y1, color='blue')
subplt1.set_title('X vs SIN(X)')

# Plot on the second subplot
subplt2.plot(x, y2, color='green')
subplt2.set_title('X vs COS(X)')

# Show the plot
plt.show()
```

```python
# Sample Data
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# import GridSpec for customizing subplots

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
gs = GridSpec(2,2)

# Create Subplots Using the GridSpec:

subplt1 = plt.subplot(gs[0, 0])  # First subplot in the first row
subplt2 = plt.subplot(gs[0, 1])  # Second subplot in the first row
subplt3 = plt.subplot(gs[1, :])  # Subplot that spans the entire second row


# Plotting data in subplots

subplt1.plot(x,y1, color = 'blue')
subplt1.set_title('X vs SIN(X)')

subplt2.plot(x,y1, color = 'green')
subplt2.set_title('X vs COS(X)')

subplt3.plot(x, y1, color='red', label='sin(x)')
subplt3.plot(x, y2, color='purple', label='cos(x)')
subplt3.legend()
subplt3.set_title('Combined Plot')

# Prevent Overlapping
plt.tight_layout()
# Show
plt.show()
```

# On Iris Dataset

```python
# Load the Iris dataset
import seaborn as sns
iris = sns.load_dataset("iris")

# Set up a 2x2 grid of subplots using GridSpec
fig = plt.figure(figsize=(12, 8))
grid = GridSpec(2, 2)

# Subplot 1: Sepal Length
subplt_1 = plt.subplot(grid[0, 0])
sns.histplot(iris, x="sepal_length", hue="species", multiple="stack", ax=subplt_1)
subplt_1.set_title("Sepal Length Distribution")

# Subplot 2: Sepal Width
subplt_2 = plt.subplot(grid[0, 1])
sns.histplot(iris, x="sepal_width", hue="species", multiple="stack", ax=subplt_2)
subplt_2.set_title("Sepal Width Distribution")

# Subplot 3: Petal Length
subplt_3 = plt.subplot(grid[1, 0])
sns.histplot(iris, x="petal_length", hue="species", multiple="stack", ax=subplt_3)
subplt_3.set_title("Petal Length Distribution")

# Subplot 4: Petal Width
subplt_4 = plt.subplot(grid[1, 1])
sns.histplot(iris, x="petal_width", hue="species", multiple="stack", ax=subplt_4)
subplt_4.set_title("Petal Width Distribution")

plt.tight_layout()
plt.show()
```
