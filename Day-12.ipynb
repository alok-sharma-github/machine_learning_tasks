# Natural Language Processing

```python
text = "Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language."
import nltk
from nltk.util import bigrams, trigrams
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, RegexpTokenizer
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk import pos_tag
```

```python
text = "In this example, the chatbot serves as an automated assistant that handles customer inquiries, providing timely and accurate responses to common questions. By implementing a chatbot, you can improve the efficiency of your customer service operations, reduce response times, and enhance the overall customer experience on your website."
```

```python
import string
from nltk.stem import WordNetLemmatizer
```

```python
tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token not in string.punctuation]

ps = PorterStemmer()
ls = LancasterStemmer()
ss = SnowballStemmer("english")
stemmed_tokens_ps = [ps.stem(token) for token in filtered_tokens]
stemmed_tokens_ls = [ls.stem(token) for token in filtered_tokens]
stemmed_tokens_ss = [ss.stem(token) for token in filtered_tokens]

pos_tags = pos_tag(filtered_tokens)


lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

print("Original Text:", text)
print("\nTokens:", tokens)
print("\nFiltered Tokens after Stopword Removal:", filtered_tokens)
print("\nStemmed Tokens PS:", stemmed_tokens_ps)
print("\nStemmed Tokens LS:", stemmed_tokens_ls)
print("\nStemmed Tokens SS:", stemmed_tokens_ss)
print("\nPOS Tags:", pos_tags)
print("Lemmatized Tokens:", lemmatized_tokens)
```

```python
def stem_words(sentence):
    # Tokenization
    tokens = word_tokenize(sentence)

    # Get English stopwords
    stop_words = set(stopwords.words('english'))

    # Remove stopwords and punctuation marks from tokens
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token not in string.punctuation]

    # Initialize stemming algorithms
    ps = PorterStemmer()
    ls = LancasterStemmer()
    ss = SnowballStemmer("english")

    # Perform stemming with each algorithm
    stemmed_tokens_ps = [ps.stem(token) for token in filtered_tokens]
    stemmed_tokens_ls = [ls.stem(token) for token in filtered_tokens]
    stemmed_tokens_ss = [ss.stem(token) for token in filtered_tokens]

    # Create a dictionary to store stemmed tokens
    stemmed_tokens = {
        "PorterStemmer": stemmed_tokens_ps,
        "LancasterStemmer": stemmed_tokens_ls,
        "SnowballStemmer": stemmed_tokens_ss
    }

    return stemmed_tokens
```

```python
def lemmatize_words(sentence):
    
    tokens = word_tokenize(sentence)
    
    stop_words = set(stopwords.words('english'))

    filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token not in string.punctuation]

    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    return lemmatized_tokens
```
