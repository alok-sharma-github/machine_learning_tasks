## Probability Distribution


Load the exam.csv dataset into a DataFrame and perform the following tasks on the “math score,” ”readingscore,” and “writing score” columns:

1.Plot the distributions using the seabornlibrary

2.Find the mean and standard deviation using the scipylibrary

3.Find skewness and kurtosisusing thestatsmodule from the scipylibrary

4.Find Standard Normal Distribution(z-score) using the statsmodule from the scipylibrary5.Plot the Z distribution using the seabornlibraryDataset:


```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:\\Users\\alok\\Desktop\\Inzint\\datasets\\StudentsPerformance.csv")

# 1. Plot the distributions using seaborn
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True, color='blue')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')

plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True, color='green')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.title('Distribution of Reading Scores')

plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True, color='red')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
plt.title('Distribution of Writing Scores')

plt.tight_layout()
plt.show()

# 2. Find the mean and standard deviation using scipy
math_mean = df['math score'].mean()
math_std = df['math score'].std()
read_mean = df['reading score'].mean()
read_std = df['reading score'].std()
write_mean = df['writing score'].mean()
write_std = df['writing score'].std()

print("Mean and Standard Deviation:")
print("Math Score: Mean - {}, Standard Deviation - {}".format(math_mean, math_std))
print("Reading Score: Mean - {}, Standard Deviation - {}".format(read_mean, read_std))
print("Writing Score: Mean - {}, Standard Deviation - {}".format(write_mean, write_std))

# 3. Find skewness and kurtosis using scipy
math_skewness = stats.skew(df['math score'])
math_kurtosis = stats.kurtosis(df['math score'])
read_skewness = stats.skew(df['reading score'])
read_kurtosis = stats.kurtosis(df['reading score'])
write_skewness = stats.skew(df['writing score'])
write_kurtosis = stats.kurtosis(df['writing score'])

print("\nSkewness and Kurtosis:")
print("Math Score - Skewness: {}, Kurtosis: {}".format(math_skewness, math_kurtosis))
print("Reading Score - Skewness: {}, Kurtosis: {}".format(read_skewness, read_kurtosis))
print("Writing Score - Skewness: {}, Kurtosis: {}".format(write_skewness, write_kurtosis))

# 4. Find Standard Normal Distribution (z-score) using scipy
math_zscore = stats.zscore(df['math score'])
read_zscore = stats.zscore(df['reading score'])
write_zscore = stats.zscore(df['writing score'])

# 5. Plot the Z distribution using seaborn
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.histplot(math_zscore, kde=True, color='blue')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Z-Scores')

plt.subplot(1, 3, 2)
sns.histplot(read_zscore, kde=True, color='green')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.title('Distribution of Reading Z-Scores')

plt.subplot(1, 3, 3)
sns.histplot(write_zscore, kde=True, color='red')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.title('Distribution of Writing Z-Scores')

plt.tight_layout()
plt.show()
```

## Sampling Distribution

A wine can be labeled as quality "7" and above only if the alcohol level is greater than 11.3%. If the level of alcohol is high, then the will taste sour enough; if the alcohol level is less than 11.3%, the wine might taste bitter. The quality controlperson is only allowed to open ten bottles froma batch of wine with quality "7" & "8"to ensure the levels are sufficient.

Load  the “winequality-red.csv”data  into  a  DataFrame  and Perform  following tasks:

1.Pick   up   a   sample   of   ten   bottles   from   "7"   and   "8"   quality   using numpy.random()function

2.Find out the mean and standard deviation of the sample

3.Plot the alcohol content available in the sample using a distplot

4.Calculate   the   standard   error   using   formula -standard_error  = s/np.sqrt(n), Where,s  =  standard  deviation  of  the  sample,  and  n  = sample size

```python
import numpy as np

df = pd.read_csv("C:\\Users\\alok\\Desktop\\Inzint\\datasets\\winequality-red.csv")

sample_size = 10
quality_7_sample = df[df['quality'] == 7].sample(n=sample_size, random_state=42)
quality_8_sample = df[df['quality'] == 8].sample(n=sample_size, random_state=42)

# 2. Find out the mean and standard deviation of the sample
quality_7_mean = quality_7_sample['alcohol'].mean()
quality_8_mean = quality_8_sample['alcohol'].mean()
quality_7_std = quality_7_sample['alcohol'].std()
quality_8_std = quality_8_sample['alcohol'].std()

print("Mean and Standard Deviation of Sample from Quality 7:")
print("Mean:", quality_7_mean)
print("Standard Deviation:", quality_7_std)
print("\nMean and Standard Deviation of Sample from Quality 8:")
print("Mean:", quality_8_mean)
print("Standard Deviation:", quality_8_std)

# 3. Plot the alcohol content available in the sample using a distplot
plt.figure(figsize=(8, 6))
sns.distplot(quality_7_sample['alcohol'], label='Quality 7')
sns.distplot(quality_8_sample['alcohol'], label='Quality 8')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Density')
plt.title('Distribution of Alcohol Content in Samples')
plt.legend()
plt.show()

# 4. Calculate the standard error
standard_error_7 = quality_7_std / np.sqrt(sample_size)
standard_error_8 = quality_8_std / np.sqrt(sample_size)

print("\nStandard Error of Sample from Quality 7:", standard_error_7)
print("Standard Error of Sample from Quality 8:", standard_error_8)
```

Our goal was to get a mean of 11.3, but we got 11.65 for our samples. Since this is a sample and we do not expect it to be precisely 11.3, it is 11.65. Does this batch  of  the  sample  accurately  reflect  the  alcohol  content  of  the  entire population? To answer these questions, perform the following tasks:

1.Take nine samples of size ten each from the population and find the means and standard deviations of all the nine samples

2.Plot the means of the sampling distribution

3.Find the standard error of the sampling distribution

```python
population = df[(df['quality'] == 7) | (df['quality'] == 8)]['alcohol']

# 1. Take nine samples of size ten each from the population and find the means and standard deviations of all the nine samples
sample_size = 10
num_samples = 9

sample_means = []
sample_stds = []

for i in range(num_samples):
    sample = population.sample(n=sample_size, random_state=i)
    sample_means.append(sample.mean())
    sample_stds.append(sample.std())

# 2. Plot the means of the sampling distribution
plt.figure(figsize=(8, 6))
sns.histplot(sample_means, kde=True, color='blue', label='Sample Means')
plt.axvline(x=np.mean(sample_means), color='red', linestyle='--', label='Mean of Sample Means')
plt.xlabel('Mean Alcohol Content (%)')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of Sample Means')
plt.legend()
plt.show()

# 3. Find the standard error of the sampling distribution
standard_error_sampling_distribution = np.std(sample_means) / np.sqrt(sample_size)

print("Standard Error of the Sampling Distribution:", standard_error_sampling_distribution)
```

## Inferential Statistics

Load the wine dataset using the Sklearn library to a DataFrame and perform the following tasks:

1.Covert the datasetinto DataFrame using pandas

2.Generate asample size of 50and givea randomstate as 100

3.Calculate Z-critical, Margin of Error & Confidence Interval for alcohol ata 95% significance interval on generated sample data

```python
import pandas as pd
from sklearn.datasets import load_wine
import numpy as np
from scipy.stats import norm

# 1. Load the wine dataset and convert it into a DataFrame
wine_data = load_wine()
df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)
df['alcohol'] = wine_data.target

# 2. Generate a sample size of 50 with a random state of 100
sample_size = 50
sample_data = df['alcohol'].sample(n=sample_size, random_state=100)

# 3. Calculate Z-critical, Margin of Error & Confidence Interval for alcohol at a 95% significance interval
confidence_level = 0.95

# Calculate the standard error
standard_error = np.std(sample_data) / np.sqrt(sample_size)

# Find the Z-critical value
z_critical = norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
sample_mean = np.mean(sample_data)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("Z-critical value:", z_critical)
print("Margin of Error:", margin_of_error)
print("Confidence Interval for alcohol at a 95% significance interval:", confidence_interval)
```

## Statistical Hypothesis Testing

A  Tyre  manufacturing  company  claims  that  their  original  Tyres  run  for  about 50,000 km before replacing them. A random sample of 100 tyres was taken, and it turned out that the average life of the sample was about 48,000 Km.To find out whether the company said the truth or lie, based on standard deviations as 1500.

H0:u  =  Tyres  run  for  about  50,000  km  before  having  to  replace  them  (Null hypothesis)

H1:  u  =  Tyres  can’t  run  for  about  50,000  km  before  having  to  replace them(Alternate hypothesis)

```python
sample_mean = 48000  # Sample mean
population_mean = 50000  # Population mean
standard_deviation = 1500  # Standard deviation
sample_size = 100  # Sample size

# Calculate the z-score
z_score = (sample_mean - population_mean) / (standard_deviation / np.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(np.abs(z_score)))  # Multiply by 2 for a two-tailed test

# Set the significance level
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis. The average lifespan of the tires is not 50,000 km.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the average lifespan of the tires is different from 50,000 km.")
```

### Problem 2:

A company  claims  that  its  employees  earn  approximately  5LPA  on  average.  A random sample of 50 employees was taken, and it turned out to have a mean of 2 LPA. Perform hypothesis testing for the given claim having a population mean of 5 LPA with a standard deviation of 3.

H0:u(Null hypothesis)  = Employees are earning on an average around 5LPA 

H1:u(Alternate hypothesis)= Employees are not earning 5LPA on an averageSample Input:Sample Output:


```python

sample_mean = 2  # Sample mean
population_mean = 5  # Population mean
standard_deviation = 3  # Standard deviation
sample_size = 50  # Sample size

# Calculate the z-score
z_score = (sample_mean - population_mean) / (standard_deviation / np.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(np.abs(z_score)))  # Multiply by 2 for a two-tailed test

# Set the significance level
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis. Employees are not earning on average around 5 LPA.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that employees are not earning on average around 5 LPA.")
```
