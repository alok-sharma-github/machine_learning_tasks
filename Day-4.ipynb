- We use GridSpec to define a 2x2 grid.
- Subplots are created using plt.subplot within this grid.
- We use sns.histplot to create stacked histograms for each feature, color-coded by species.
- Titles are set for each subplot to describe the distribution being visualized.

```python
# subplot()
# GridSpec()
'''
create a grid of subplots to visualize relationships between different 
features of the Iris dataset. Specifically, compare the 
distributions of sepal length, sepal width, petal length, and petal 
width for the three different species of
iris flowers (setosa, versicolor, and virginica)
'''
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Set up a 2x2 grid of subplots using GridSpec
fig = plt.figure(figsize=(12, 8))
grid = GridSpec(2, 2, width_ratios=[1, 1], height_ratios=[1, 1])

# Subplot 1: Sepal Length
ax0 = plt.subplot(grid[0, 0])
sns.histplot(iris, x="sepal_length", hue="species", multiple="stack", ax=ax0)
ax0.set_title("Sepal Length Distribution")

# Subplot 2: Sepal Width
ax1 = plt.subplot(grid[0, 1])
sns.histplot(iris, x="sepal_width", hue="species", multiple="stack", ax=ax1)
ax1.set_title("Sepal Width Distribution")

# Subplot 3: Petal Length
ax2 = plt.subplot(grid[1, 0])
sns.histplot(iris, x="petal_length", hue="species", multiple="stack", ax=ax2)
ax2.set_title("Petal Length Distribution")

# Subplot 4: Petal Width
ax3 = plt.subplot(grid[1, 1])
sns.histplot(iris, x="petal_width", hue="species", multiple="stack", ax=ax3)
ax3.set_title("Petal Width Distribution")

plt.tight_layout()
plt.show()
```
