#Problem Statement 1:

Data Filtering using Pandas

Write a Python program that reads the player.csv  file into a DataFrame, then shows only those players who played more than 150 innings and scored at least two centuries and 35 fifties.

```python
# Import the pandas library and alias it as pd
import pandas as pd

# Read the CSV file into a DataFrame
players = pd.read_csv("C:\\Users\\alok\\Downloads\\player.csv")

# Display the first few rows of the DataFrame
players.head()

# Filter players based on the given conditions
filtered_players = players[
    (players['innings'] > 150) &  # Players who played more than 150 innings
    (players['100s'] >= 2) &      # Players who scored at least two centuries
    (players['50s'] >= 35)        # Players who scored at least 35 fifties
]

# Display the filtered DataFrame
print(filtered_players)
```

Problem Statement 4: Combining DataFrames -

Write  a  Python  programto  create  two  DataFrames.  First  DataFrame  with columns “employee”and “department,”Second  DataFrame  with  columns “employee”and “joining_date,” and Perform the following tasks:

1.Merge the DataFrames based on the “employee”column

2.Display the final DataFrame with employees records who are working in the “Engineering” department

```python
# Sample DataFrames
df_department = pd.DataFrame({
    'employee': [101, 102, 103, 104],
    'department': ['Engineering', 'HR', 'Marketing', 'Finance']
})

df_joining_date = pd.DataFrame({
    'employee': [101, 102, 103, 104],
    'joining_date': ['2021-01-15', '2022-03-05', '2021-05-20', '2023-02-10']
})

# Merge DataFrames based on the "employee" column
merged_df = pd.merge(df_department, df_joining_date, on='employee')

# Display the final DataFrame with employees in the "Engineering" department
engineering_df = merged_df[merged_df['department'] == 'Engineering']
print(engineering_df)
```

Load the “sample_superstore.csv”dataset into a DataFrame; using the Seaborn library, plot the following insights from the dataset:

1.Print a correlation matrix between Postalcode,Quantity,Sales,Discount,andProfit

2.Plot the correlation matrix using a heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv("C:\\Users\\alok\\Downloads\\SampleSuperstore.csv")

# column names
df.columns


```

```python
# 1. Print a correlation matrix
correlation_matrix = df[['Postal Code', 'Quantity', 'Sales', 'Discount', 'Profit']].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# 2. Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

 Problem Statement: Sales Data Analysis-I
 
 Load the “sample_superstore.csv”dataset provided  a DataFrame, using the Matplotlib library, plot the following insights from the dataset:
 
 1.Compare the total sales for each category of products
 
 2.Number of deals in each state of America
 
 Note:You can use any plot of your choice to show the insights

```python
# 1. Compare the total sales for each category of products
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plot the total sales for each category
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales for Each Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()

# 2. Number of deals in each state of America
state_deals = df['State'].value_counts()

# Plot the number of deals in each state
plt.figure(figsize=(12, 6))
state_deals.plot(kind='bar', color='lightcoral')
plt.title('Number of Deals in Each State of America')
plt.xlabel('State')
plt.ylabel('Number of Deals')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 Problem Statement: Data Visualization using MatplotlibLoad 
 
 load the “Netflix_titles.csv” data  into  a  DataFrame  and  plot  the  following insights using the Matplotliblibrary:
 
 1.How much of the Netflix content are TV shows or Movies?
 
 2.In which country is Netflix the most popular?

```python
# Load the dataset into a DataFrame
df = pd.read_csv("C:\\Users\\alok\\Downloads\\netflix_titles.csv")

# columns
df.columns


```

```python
# 1. How much of the Netflix content are TV shows or Movies?
content_distribution = df['type'].value_counts()

# Plot the distribution of TV shows and Movies
plt.figure(figsize=(8, 5))
content_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Distribution of Netflix Content (TV Shows vs. Movies)')
plt.ylabel('')
plt.show()

# 2. In which country is Netflix the most popular?
popular_countries = df['country'].value_counts().head(10)

# Plot the top 10 countries where Netflix is most popular
plt.figure(figsize=(12, 6))
popular_countries.plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Countries Where Netflix is Most Popular')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
